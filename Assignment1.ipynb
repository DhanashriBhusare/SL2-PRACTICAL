```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
x

def sigmoid(x):
	return 1 / (1 + np.exp(-x))

def relu(x):
	return np.maximum(0, x)

def tanh(x):
	return np.tanh(x)

def leaky_relu(x, alpha=0.01):
	return np.where(x > 0, x, x * alpha)

def elu(x, alpha=1.0):
	return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

def softmax(x):
	e_x = np.exp(x - np.max(x))
	return e_x / e_x.sum(axis=0)




```

```python
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
y_leaky_relu = leaky_relu(x)
y_elu = elu(x)
y_softmax = softmax(x)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create input data
x = np.linspace(-5, 5, 100)

# Initialize variables for activation functions
y_sigmoid = None
y_relu = None
y_tanh = None
y_leaky_relu = None
y_elu = None
y_softmax = None

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# Menu-driven program
while True:
    print("\nActivation Function Visualization Menu:")
    print("1. Sigmoid")
    print("2. ReLU")
    print("3. Tanh")
    print("4. Leaky ReLU")
    print("5. ELU")
    print("6. Softmax")
    print("0. Exit")

    choice = input("Enter your choice (0-6): ")

    if choice == '0':
        print("Exiting the program. Goodbye!")
        break
    elif choice == '1':
        y_sigmoid = sigmoid(x)
        plt.plot(x, y_sigmoid, label='Sigmoid')
        plt.title('Sigmoid')
    elif choice == '2':
        y_relu = relu(x)
        plt.plot(x, y_relu, label='ReLU')
        plt.title('ReLU')
    elif choice == '3':
        y_tanh = tanh(x)
        plt.plot(x, y_tanh, label='Tanh')
        plt.title('Tanh')
    elif choice == '4':
        y_leaky_relu = leaky_relu(x)
        plt.plot(x, y_leaky_relu, label='Leaky ReLU')
        plt.title('Leaky ReLU')
    elif choice == '5':
        y_elu = elu(x)
        plt.plot(x, y_elu, label='ELU')
        plt.title('ELU')
    elif choice == '6':
        y_softmax = softmax(x)
        plt.plot(x, y_softmax, label='Softmax')
        plt.title('Softmax')
    else:
        print("Invalid choice. Please enter a number between 0 and 6.")

    plt.show()
```
